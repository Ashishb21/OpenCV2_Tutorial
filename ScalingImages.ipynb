```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
# Scaling [resizing the images, scaling down ]
# zooming [ scaling up] Interpoltaion Methods
# Transformation of the images using Affine methods 
# Rotation of the images 
# Applying filters 
```

```python
# Image resize
img1 = cv2.imread('./misc/4.1.02.tiff', 1)
output = cv2.resize( img1, None, fx=0.5, fy=1.5, interpolation=cv2.INTER_AREA )
cv2.imshow('Output', output)
cv2.waitKey(0)
cv2.destroyAllWindows(
```

```python
# Transformation of  image 
img1 = cv2.imread('./misc/4.1.02.tiff', 1)
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
columns,rows,channels = img1.shape
T= np.float32([[1, 0, 150], [0, 1, 50]])  # transfer x=150 n y =50
output = cv2.warpAffine(img1, T, (columns, rows))
plt.imshow(output)
plt.show()
```

```python
# Rotation Matrix 

img1 = cv2.imread('./misc/4.1.02.tiff', 1)
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
columns,rows,channels = img1.shape
R = cv2.getRotationMatrix2D((columns/2, rows/2), 45, 1)  # 45 degree rotation , scaling factor =1 [applied on both dim]
print(R)
output = cv2.warpAffine(img1, R, (columns, rows))
plt.imshow(output)
plt.show()
```

```python
#Applying filters
img1= cv2.imread('./misc/4.1.02.tiff',1)
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
plt.imshow(img1)
k = np.array(np.ones((11, 11), np.float32))/121
k = np.array(([0, -1, 0], [-1, 5, -1], [0, -1, 0]), np.float32)   
print(k)
output = cv2.filter2D(img1, -1, k)

plt.subplot(1, 2, 1)
plt.imshow(img1)
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(output)
plt.title('Filtered Image')

plt.show()
print (output.shape)
```
