```python
# Read an image --
imgpath='./misc/4.1.01.tiff'
img=cv2.imread(imgpath)
cv2.imshow('lena',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Read an image with grey scale --
imgpath='./misc/4.1.04.tiff'
outpath ='/output/4.1.04.tiff'
img=cv2.imread(imgpath,0) # specificy 0 
cv2.imshow('lena',img)
cv2.imwrite(outpath,img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# read an image and display using matplotlib
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
imgpath='./misc/4.1.08.tiff'
img1= cv2.imread(imgpath,0)


plt.imshow(img1,cmap='gray')   # cmap cmap='grey' shows lists of cmap
plt.show()

plt.imshow(img1)
plt.title('Default Colormap')
plt.xticks([])
plt.yticks([]) # rulers dissapear
plt.show()  
```

```python
#displaying the image using colormap BGR to RGB 

import cv2
import matplotlib.pyplot as plt
%matplotlib inline
imgpath='./misc/4.1.08.tiff'
img= cv2.imread(imgpath)
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img1)   # cmap cmap='grey' shows lists of cmap
plt.show()

plt.imshow(img)
plt.title('Default Colormap')
plt.xticks([])
plt.yticks([]) # rulers dissapear
plt.show()  
```
