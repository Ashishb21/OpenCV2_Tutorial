```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import time 
# Addition subtration division multplication of 2 images 
# blending using add weighted and transition of 2 images 
#Trackbar demo for blending and trasition of the images 
# splitting of the rgb values and then merge
# Finding the negative of the colored images and grey scale image [255-img]
```

```python
#Addition 
img1 = cv2.imread('./misc/4.1.01.tiff')
img2= cv2.imread('./misc/4.1.03.tiff')
img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
img3=img1+img2
img=[img1,img2,img3]
for i in range(0,3):
    plt.subplot(1,3,i+1)
    plt.imshow(img[i])
plt.show()
```

```python
#Multiplication
img1=cv2.imread('./misc/nancy.jpg')
img1= cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2= img1*img1
plt.subplot(1,2,2)
plt.imshow(img1)
plt.subplot(1,2,1)
plt.imshow(img2)
plt.show()
```

```python
# blending and transition of the images .

img1=cv2.imread('./misc/4.1.01.tiff')
img2=cv2.imread('./misc/4.1.04.tiff')
for i in np.linspace(0, 1, 100):
        alpha = i
        beta = 1 - alpha #[alpha + beta =1]
        output = cv2.addWeighted(img1, alpha, img2, beta, 0)
        cv2.imshow('Transition', output)
        time.sleep(0.02)
        if cv2.waitKey(1) == 27:
            break
        
cv2.destroyAllWindows()
```

```python
# track bar for blending and transition of the images 

img1=cv2.imread('./misc/4.1.01.tiff')
img2=cv2.imread('./misc/4.1.04.tiff')

output = cv2.addWeighted(img1, 0.5, img2, 0.5, 0)    
windowName = "Transition Demo"
cv2.namedWindow(windowName)
    
cv2.createTrackbar('Alpha', windowName, 0, 1000, emptyFunction)
    
while(True):
        
    cv2.imshow(windowName, output)

    if cv2.waitKey(1) == 27:
        break

    alpha = cv2.getTrackbarPos('Alpha', windowName) / 1000
    beta = 1 - alpha

    output = cv2.addWeighted(img1, alpha, img2, beta, 0)

    print (alpha, beta)
        
cv2.destroyAllWindows()
```

```python
# split and merge of the RGB 
img1=cv2.imread('./misc/4.2.01.tiff')
img1= cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
r,g,b=cv2.split(img1)
img3= cv2.merge((r,g,b))
plt.subplot(2,5,2)
plt.imshow(r,cmap='Reds')
plt.subplot(2,5,1)
plt.imshow(g,cmap='Greens')
plt.subplot(2,5,3)
plt.imshow(b,cmap='Blues')
plt.subplot(2,5,4)
plt.imshow(img3)
plt.subplot(2,5,5)
plt.imshow(img3)
plt.show()
```
