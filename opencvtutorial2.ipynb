```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

# capture still images 
# change the resolution to custom values of the webcam
# save the video to a file 
```

```python
#Capture still images 

cap=cv2.VideoCapture(0) # can attach multiple web cams 
                       #cap1= cv.VideoCapture(1)
if cap.isOpened():
    ret, frame = cap.read()
else:
    ret = False

img1 = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    
plt.imshow(img1)
plt.title('Color Image RGB')
plt.xticks([])
plt.yticks([])
plt.show()
cap.release()
    
```

```python
#Live video feed with gray and color image 
windowName = "Live Video Feed"
cv2.namedWindow(windowName)
cap = cv2.VideoCapture(0)
    
if cap.isOpened():
    ret, frame = cap.read()
else:
    ret = False

while ret:
    
    ret, frame = cap.read()
        
    output = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        
    cv2.imshow("Gray", output)
    cv2.imshow(windowName, frame)
    if cv2.waitKey(1) == 27:
        break

cv2.destroyAllWindows()    

cap.release()
```

```python
#Setting the resolution of the webcam

windowName = "Live Video Feed"
cv2.namedWindow(windowName)
cap = cv2.VideoCapture(0)
    
print('Width : ' + str(cap.get(3)))
print('Height : ' + str(cap.get(4)))
    
cap.set(3, 5000)
cap.set(4, 5000)
    
print('Width : ' + str(cap.get(3)))
print('Height : ' + str(cap.get(4)))
    
if cap.isOpened():
    ret, frame = cap.read()
else:
    ret = False

while ret:
    
    ret, frame = cap.read()
        
    output = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        
    cv2.imshow("Gray", output)
    cv2.imshow(windowName, frame)
    if cv2.waitKey(1) == 27:
        break

cv2.destroyAllWindows()    

cap.release()
```

```python
#save the video feed to a file 
windowName = "Live Webcam Video Feed Capture"
cv2.namedWindow(windowName)
    
cap = cv2.VideoCapture(0)
    
filename = './output/out.avi'
codec = cv2.VideoWriter_fourcc('W', 'M', 'V', '2')
framerate = 30
resolution = (640, 480)
    
VideoFileOutput = cv2.VideoWriter(filename, codec, framerate, resolution)
    
if cap.isOpened():
    ret, frame = cap.read()
else:
    ret = False

while ret:
    
    ret, frame = cap.read()
        
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
        
    VideoFileOutput.write(frame)
        
    cv2.imshow(windowName, frame)
    if cv2.waitKey(1) == 27:
        break

cv2.destroyAllWindows()    
VideoFileOutput.release()
cap.release()
```
